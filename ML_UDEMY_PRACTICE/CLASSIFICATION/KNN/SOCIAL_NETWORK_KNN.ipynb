```python
from sklearn.model_selection import train_test_split as tts
from sklearn.preprocessing import StandardScaler as sc
from sklearn.neighbors import KNeighborsClassifier
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

```python
print(x,y)
```

```python
x_train,x_test,y_train,y_test=tts(x,y,test_size=0.2,random_state=0)
```

```python
print(x_train.shape,x_test.shape,y_train.shape,y_test.shape)
```

```python
model=KNeighborsClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
y_prev=model.predict(x_test)
```

```python
print(y_prev)
```

```python
# model.score(y_prev.reshape(-1,1),y_test)
```

```python
import sklearn.metrics 
```

```python
print(sklearn.metrics.classification_report(y_prev,y_test))
```

```python
print(sklearn.metrics.confusion_matrix(y_prev,y_test))
```

```python
error_rate=[]
for i in range(1,40):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    y_i=knn.predict(x_test)
    print(f'y_i{y_i} \n, y_test: {y_test}')
    print('-'*125)
    print(f'{np.mean(y_i!=y_test)}')
    error_rate.append(np.mean(y_i!=y_test))
print(f'Minimum error_rate is {min(error_rate)} and index/k={error_rate.index(min(error_rate))}' )
```
